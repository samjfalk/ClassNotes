## The Dataset

The data is in a csv file in your repo. It contains the following columns:

- **world_rank** - world rank for the university. Contains rank ranges and equal ranks (e.g. = 94 and 201-250).
- **university_name** - name of university.
- **country** - country of each university.
- **teaching** - university score for teaching (the learning environment).
- **international** - university score international outlook (staff, students, research).
- **research** - university score for research (volume, income and reputation).
- **citations** - university score for citations (research influence).
- **income** - university score for industry income (knowledge transfer).
- **total_score** - total score for university, used to determine rank.
- **num_students** - number of students at the university.
- **student_staff_ratio** - Number of students divided by number of staff.
- **international_students** - Percentage of students who are international.
- **female_male_ratio** - Female student to Male student ratio.
- **year** - year of the ranking (2011 to 2016 included).

The target is the **total score**, which directly corresponds to the final ranking.

**N.B. - if the total score is reported as "-" that will be considered a 0 when scoring.**

```python
df['country'].replace(to_replace='Unisted States of America ', value='United States of America', inplace=True)
```

```python
def nana(x):
    if x=='-':
        x=np.nan
        return x
    else:
        return x


df=df.applymap(nana)
```

```python
def students(x):
   try:
       return (float(x.split(',')[0]))*1000+(float(x.split(',')[1]))
   except:
       return x
       
df['num_students'] = df['num_students'].apply(students)
```

```python
df['total_score'] = (df['total_score'].astype(float))
df['international'] = (df['international'].astype(float))
df['income'] = (df['income'].astype(float))
```

```python
pair = sns.PairGrid(df)
pair = pair.map(plt.scatter)
```

```python
"""df['female_male_ratio'] = (df['female_male_ratio'].replace( ':00','', regex=True ))
df['female_male_ratio'] = (df['female_male_ratio'].replace( ' : ','', regex=True ))
df['female_male_ratio'] = (df['female_male_ratio'].replace( ':','', regex=True ))


i=0
for y in df['female_male_ratio']:
   if y > 0:
       if y == '100':
           df['female_male_ratio'][i] = np.nan
       else:
           df['female_male_ratio'][i] = float(df['female_male_ratio'][i][0:2])/float(df['female_male_ratio'][i][2:4])
   i+=1
```

```python
def p2f(x):
   try:
       return float(x.replace('%',''))/100
   except:
       return x
       

df['international_students'] = df['international_students'].apply(p2f)
```

```python
test_data=df.loc[df['year'] == 2016]
train_data=df.loc[df['year'] != 2016]
```

```python
test_data.dropna(inplace=True)
train_data.dropna(inplace=True)
```

```python
sns.heatmap(df.corr())
```

```python
X_train= train_data[['teaching', 'research']]
y_train=train_data['total_score']
X_test= test_data[['teaching', 'research']]
y_test=test_data['total_score']
```

```python
lr=linear_model.LinearRegression()
lr_model=lr.fit(X_train,y_train)
predictions=lr_model.predict(X_train)
```

```python
ax=sns.regplot(predictions,y_train)
```

```python
lr_r2 = r2_score(y_train, predictions)
lr_r2
```

```python
len(lr_model.coef_)
lr_model.coef_
```

```python
lasso=linear_model.Lasso(alpha=1)
lasso_model=lasso.fit(X_train, y_train)
```

```python
y_true_lasso = y_train
y_pred_lasso = lasso_model.predict(X_train)
lasso_r2 = r2_score(y_true_lasso, y_pred_lasso)
lasso_r2
```

In this lab today, we will run the kNN model to determine the best ‘k’ value in order to determine whether a wine is of high quality. 

Use the following features: ‘density’, ‘sulphates’, residual sugar’.

Please check 'k' values from 1 to 50 in order to determine the best value.

Below is starert code which obtains demo data from Amazon Web Services website:

```python
import numpy as np
import pandas as pd
import pylab as pl
from sklearn.neighbors import KNeighborsClassifier
from sklearn import datasets
from sklearn.cross_validation import train_test_split
import statistics
from statistics import mode
from sklearn.metrics import mean_squared_error
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv("https://s3.amazonaws.com/demo-datasets/wine.csv")

test_idx = np.random.uniform(0, 1, len(df)) <= 0.3
train = df[test_idx==True]
test = df[test_idx==False]

features = ['density', 'sulphates', 'residual_sugar']
```

```python
x_train=zip(zip(train[features[0]], train[features[1]], train[features[2]]), y_train)
x_train
```

```python
x_test=zip(test[features[0]], test[features[1]], test[features[2]])
x_test
```

```python
y_test= test['high_quality']
y_test
```

```python
def calculate_distance(p, q):
    distance= (((p[0]-q[0])**2)+((p[1]-q[1])**2)+((p[2]-q[2])**2))**.5
    return distance
```

```python

def knn_classifier(labeled_points, new_point, k):
    distances_points=[]
    if k%2!=0:
        for x in labeled_points:
            distances_points.append((calculate_distance(x[0], new_point),x[1]))

        distances_points.sort()
        prediction_list=[class_return[1] for class_return in distances_points[:k]]

       
        prediction=mode(prediction_list)
        """ Given classified points and one new point, return a prediction based on the votes of k neighbors."""
        return prediction
    else:
        k=k+1
        return knn_classifier(labeled_points, new_point, k)
```

```python
knn_classifier(x_train, (0.99719999999999998, 0.55000000000000004, 1.8999999999999999), 8)
```

```python
qual_dict={}
for i in range(3,30):
    qual_lst=[]
    for x in x_test:
        quality= knn_classifier(x_train, x, i)
        qual_lst.append(quality)
    qual_dict['kN{0}'.format(i)]=qual_lst
        
```

```python
MSEt=[]
for k in qual_dict:
    MSEt.append((k, mean_squared_error(y_test, qual_dict[k])))

print MSEt
```
